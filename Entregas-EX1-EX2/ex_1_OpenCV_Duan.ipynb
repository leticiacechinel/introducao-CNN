```python
#--- Lendo imagem
img = cv.imread('figs/messi5.jpg')

#--- Plotando imagem
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
#--- Fazendo crop da bola
ball = img[287:337, 337:387].copy()
plt.imshow(cv.cvtColor(ball, cv.COLOR_BGR2RGB))
```

```python
#--- Fazendo crop da cabea
head = img[80:130, 212:262].copy()
plt.imshow(cv.cvtColor(head, cv.COLOR_BGR2RGB))
```

```python
#--- Fazendo crop do espao vazio
vazio = img[280:330, 120:170].copy()
plt.imshow(cv.cvtColor(vazio, cv.COLOR_BGR2RGB))
```

```python
#--- Bola por Cabea
img[287:337, 337:387], img[80:130, 212:262] = head, ball
```

```python
#--- Plotando imagem
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
#--- Colocando a bola em outro lugar
img2 = cv.imread('figs/messi5.jpg')

img2[287:337, 100:150] = ball
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
```

```python
#--- colocando espao vazio no lugar da bola
img2[287:337, 337:387] = vazio
```

```python
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
```
